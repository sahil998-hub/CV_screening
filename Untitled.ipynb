# Cleaning Data:
1. URL
2. hash
3. mentions
4. special letters
5. punctuation

```python
import re
def cleanResume(txt):
#     cleanTxt=re.sub('http\S+\s',' ',txt)
    cleanTxt=re.sub('http\S+',' ',txt)
    cleanTxt=re.sub('@\S+',' ',cleanTxt)
    cleanTxt=re.sub('#\S+',' ',cleanTxt)
    cleanTxt=re.sub('RT|cc',' ',cleanTxt)
    cleanTxt=re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ',cleanTxt)
    cleanTxt=re.sub(r'[^\x00-\x7f]',' ',cleanTxt)
    cleanTxt=re.sub('\s+',' ',cleanTxt)
    
    return cleanTxt
```

```python
cleanResume("hello my name is http://hellowor @gmail shashank #####hrishap naman # % * ")
```

```python
df['Resume']=df['Resume'].apply(lambda x: cleanResume(x))
df
```

# Word into categorical value

## Vactorization

### Words into categorical value


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category']=le.transform(df['Category'])
```

```python
df
```

```python
df.Category.unique()
```

# Vactorization

```python
# df['Resume'][0]
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requiredText=tfidf.transform(df['Resume'])
```

```python
df
```

# Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(requiredText,df['Category'],test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Now lets train the model and print the classifiation report:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

clf=OneVsRestClassifier(LogisticRegression())
clf.fit(x_train,y_train)
ypred=clf.predict(x_test)
print(accuracy_score(y_test,ypred))
```

```python
import pickle
pickle.dump(tfidf,open('tfidf.pkl','wb'))
pickle.dump(clf,open('clf.pkl','wb'))
```
